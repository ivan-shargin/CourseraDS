# Распределение Коши(Cauchy)

Распределение Коши имеет плотность f(x) = 1/ (pi * (1 + x^2)). 
Это распределение интересно тем, что для него нельзя определить математическое ожидание.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

Сгенерируем выборку обьема 1000 из распределения Коши 

```python
cauchy = sts.cauchy()
sample = cauchy.rvs(1000)
```

Гистограмма выборки:

```python
# histogram
plt.hist(sample, density=True, label='histogram')
plt.ylabel('fraction of samples, pdf')
plt.xlabel('x')

# teoretical PDF
x = np.linspace(-200, 200, 1000)
pdf = cauchy.pdf(x)
plt.plot(x, pdf, label='teoretical PDF')
plt.ylim([0, 0.02])
plt.xlim([-200, 200])
plt.legend(loc='upper left')
```
