# Фазовое расслоение. Решетчатая жидкость.


```python
from scipy import misc
import numpy as np
w = np.arange(0.01, 1, 0.01)

#print(s)
#print(w)
def F(w, phlori):
    s = 1 - w
    return w * np.log(w) + s * np.log(s) + phlori * s * w

print(F(w, 0.25))
```

```python
%matplotlib inline
```

```python
from matplotlib import pylab as plt
```

```python
plt.plot(w, F(w, 2.5))
```

```python
def get_color():
    for item in ['r', 'g', 'b', 'c', 'm', 'y', 'k']:
        yield item



color = get_color()


phlori = 2.5
phlori_step = 0.1

for group in range(7):
    loop_phlori = phlori + phlori_step * group
    acolor = next(color)
    
    #plt.scatter(w, F(w, loop_phlori) , color=acolor, marker='o')
    plt.plot(w, F(w, loop_phlori), acolor, label = str(group))

plt.legend()
plt.xlim((0, 1))
#plt.show()
```
