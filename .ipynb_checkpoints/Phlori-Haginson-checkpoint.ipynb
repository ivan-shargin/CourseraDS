# Фазовое расслоение. Решетчатая жидкость.


```python
from scipy import misc
import numpy as np
w = np.arange(0.01, 1, 0.01)

#print(s)
#print(w)
def F(w, phlori):
    s = 1 - w
    return w * np.log(w) + s * np.log(s) + phlori * s * w

print(F(w, 0.25))
```

```python
from matplotlib import pylab as plt
```

```python
plt.plot(w, F(w, 2.5))
```
