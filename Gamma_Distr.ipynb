```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
```

# Гамма распределение.

Гамма распределение имеет плотность:  

$$f(x) = x^{a-1} * exp(-x)/Г(a) $$ где a - параметр, Г(а) - гамма функция

```python
a = 2.5 # parameter of Gamma distribution
gamma = sts.gamma(a)
```

## 1. Выборка обьема 1000

 Сгенерируем выборку обьема 1000 из Гамма распределения

```python
sample = gamma.rvs(1000)
```

Гистограмма для выборки и теоретическая плотность

```python
#histogram
plt.hist(sample, density=True, label='histogram')

#teoretical pdf
x = np.linspace(0, 10, 100)
pdf = gamma.pdf(x)
plt.plot(x, pdf, label='teoretical PDF')

plt.ylabel('$frequences$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

## 2. Выборки обьема 5, 10 и 50

Сгенерируем массивы для 1000 выборок обьема 5, 10 и 50

```python
arr_sample_5 = [gamma.rvs(5) for i in range(1000)]
arr_sample_10 = [gamma.rvs(10) for i in range(1000)]
arr_sample_50 = [gamma.rvs(50) for i in range(1000)]
```

Рассчитаем теоретические математические ожидания и среднеквадратичные отклонения. 
Для Гамма распределения с параметром a

Математическое ожидание E = a = 2.5;

Дисперсия D = a = 2.5; среднеквадратичное отклонение: $$\sqrt{a}$$

Тогда распределение выборочных средних оценивается нормальным распределением со средним a

Среднеквадратичным отклонением $$\sqrt{D} = \sqrt{D/N} = \sqrt{a/N}$$

где N - обьем выборки

```python
mean_sample_5 = [sample.mean() for sample in arr_sample_5]
mean_sample_10 = [sample.mean() for sample in arr_sample_10]
mean_sample_50 = [sample.mean() for sample in arr_sample_50]
```

1. Для выборок обьема 5:

```python
plt.hist(mean_sample_5, density=True, label='histogram for N=5')

norm_distr = sts.norm(a, np.sqrt(a/5))
x = np.linspace(0, 5, 1000)
pdf = norm_distr.pdf(x)
plt.plot(x, pdf, label='PDF for normal distr')

plt.ylabel('frequences')
plt.xlabel('mean values')
plt.legend(loc='upper right')
```

2. Для выборок обьема 10

```python
plt.hist(mean_sample_10, density=True, label='histogram for N=10')

norm_distr = sts.norm(a, np.sqrt(a/10))
x = np.linspace(0, 5, 1000)
pdf = norm_distr.pdf(x)
plt.plot(x, pdf, label='PDF for normal distr')

plt.ylabel('frequences')
plt.xlabel('mean values')
plt.legend(loc='upper right')
```

3. Для выборок обьема 50:

```python
plt.hist(mean_sample_50, density=True, label='histogram for N=50')

norm_distr = sts.norm(a, np.sqrt(a/50))
x = np.linspace(0, 5, 1000)
pdf = norm_distr.pdf(x)
plt.plot(x, pdf, label='PDF for normal distr')

plt.ylabel('frequences')
plt.xlabel('mean values')
plt.legend(loc='upper right')
```

## Выводы:

Точность аппроксимации распределения выборочных средних нормальным с ростом N увеличивается. 
С ростом N аппроксимирующее нормальное распределение становится все больше похоже на гистограмму выборочных средних.
Cреднеквадратичное отклонение обратно пропорционально корню из обьема выборки N.
